```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files  # this lets you upload files

def get_contours(img, lower_color, upper_color, min_area=500):
    hsv_img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
    mask = cv2.inRange(hsv_img, lower_color, upper_color)
    contours, _ = cv2.findContours(mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    filtered_contours = [c for c in contours if cv2.contourArea(c) >= min_area]
    return filtered_contours, mask
```

```python
uploaded = files.upload()
```

```python
img = cv2.imread('airplane.png')  # use the exact file name you uploaded

lower_airplane = np.array([0, 0, 20])
upper_airplane = np.array([90, 255, 150])
```

```python
contours, mask = get_contours(img, lower_airplane, upper_airplane)

# Draw the contours
output_img = img.copy()
cv2.drawContours(output_img, contours, -1, (0, 0, 255), 2)

# Display
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.title('Mask')
plt.imshow(mask, cmap='gray')
plt.axis('off')

plt.subplot(1,2,2)
plt.title('Contours')
plt.imshow(cv2.cvtColor(output_img, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```
